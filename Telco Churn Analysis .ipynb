```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x = 'SeniorCitizen',data = df,hue = 'Churn')
ax.bar_label(ax.containers[0])

plt.title( "count of senior citizen")
```

```python
percentage_df = (
    df.groupby(['SeniorCitizen', 'Churn']).size()
    .groupby(level=0)
    .apply(lambda x: 100 * x / float(x.sum()))
    .unstack()
)

# Plot stacked bar chart
percentage_df.plot(
    kind='bar',
    stacked=True,
    figsize=(5, 5),
    color=['skyblue', 'salmon']
)

# Add percentage labels
for i, bar_group in enumerate(percentage_df.values):
    for j, value in enumerate(bar_group):
        plt.text(
            i, 
            percentage_df.iloc[i, :j].sum() + value / 2, 
            f"{value:.1f}%", 
            ha='center', va='center', color='white', fontsize=10
        )

# Customize plot
plt.title("Percentage of Senior Citizens by Churn")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn", loc='upper right')
plt.show()
```

### from above charte we get there is no impact  of senior citizens and  gender

```python


# List of columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots
n_cols = 3  # Number of columns in the subplot grid
n_rows = -(-len(columns) // n_cols)  # Calculate the number of rows needed
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10), constrained_layout=True)

# Flatten the axes for easy indexing
axes = axes.flatten()

# Plot each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue = "Churn")
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

plt.show()
```

#### Services like OnlineSecurity, TechSupport, and DeviceProtection show higher churn rates among customers who do not use or opt out of these services. Similarly, customers without streaming services (TV or movies) exhibit a greater likelihood of churn. On the other hand, PhoneService and MultipleLines have slightly lower churn rates for customers with no phone service or no additional lines.

# Conclusion
* Customers who do not subscribe to value-added services (e.g., OnlineSecurity, TechSupport, and DeviceProtection) are more likely to churn.
* Churn rates are higher among customers with month-to-month contracts compared to those with longer-term agreements.
* Fiber optic users exhibit higher churn rates compared to DSL or no internet services, likely due to service quality or pricing issues.
* Streaming services (TV and movies) seem to contribute positively to customer retention, as churn is lower among their users.
